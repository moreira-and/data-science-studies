## CRISP-DM - Cross Industry Standard Process for Data Mining

P: O que é? R: Ciclo de desonvolvimento (metodologia) para organizar e guiar projetos em ciência e mineiração de dados dividindo-os em 6 etapas principais:
1. Business Undertanding
    - Compreender o problema ou objetivo do negócio (estimular perguntas ao cliente sobre quais são as dores, quais hipóteses queremos testar, quais perguntas queremos responder).
        - Quais são as pessoas de negócios? Quem são os stakeholdes deste projeto? O que elas querem de verdade? (Ficar próxima da área de negócio). P.A. Identificar as personas e caracterizá-las para entender o contexto.
2. Data Understanding
    - Coletar e explorar os dados disponíveis (Buscar com as personas responsáveis por armazenar os dados. Onde eu busco a informação? Quem é o detendor do dado? Onde está a base de clientes? Quais os dados que eu vou buscar nestas tabelas?) (Web Scraping, API, CSV, JSON)
3. Data Preparation
    - Preparação e organização de dados. (Limpeza, transformação, feature engeniering, CAED)
4. Modeling
    - Aplicar técnicas de machine learning para resolver o problema. (Regressão, Classificação, Agrupamento)
5. Evaluation
    - Verificar se o modelo atende aos objetivos do negócio. (O resultado está fazendo sentido? Aplicar técnicas de avaliação do projeto, validar com a área de negócio)
6. Deployment
    - Colocar o modelo em produção para gerar valor. 

- Estrutura clara e organizada desde o início ao fim, com flexidade de aplicações em diverentes projetos, para diversa áreas de maneira interativa, permitindo revisitar as etapas para fazer ajustas e melhorias e focada no negócio pois começa com a compreensão dos objetivos do negócio para garantir alinhamento com as necessidades reais. Permite reusabilidade dos templates e é amplamente aceito no mercado. Além disso, promove uma redução de riscos e falhas dos resultados finais ao identificar e resolver problemas nas primeiras fases.


## Modularização em Python

- Modularização = Boa prática para criar código mais organizado, reunitlizável e fácil de manter
    - Significa dividir o projeto em partes menores e independentes (módulos), cada uma responsável por uma funcionalidade específica.
    - Reunilização de código em diversos projetos.
    - Projeto mais estruturado e fácil de navegar
    - Facilidade de manutenção e correção de problemas (principalmente, em escalas)
    - Maior testabilidade com módulos independentes
    - Aprender a criar rotinas bem estruturadas para reduzir o nível de dificuldade do código
    - Facilita a colaboração em equipe, pois cada desenvolvedor pode trabalhar em diferentes módulos, sem interferir diretamente no código de outros.
    - Escalabilidade, pois permite que o projeto cresça de maneira organizada
    - Redução de dependências entre diferentes partes do código, deixando o código resiliente a mudanças.

## Pacotes

- É um diretório que contém vários módulos (arquivos.py) e geralmente um arquivo __init__.py (mesmo vazio) para indicar que é um pacote.
    - permite organizar melhor o código e agrupar funcionalidades relacionadas

    

----- __ init__.py ---- 
from calculadora.operacoes_basicas import soma

## Boas práticas para modularização

- Separar responsabilidades
    - Cada módulo deve ser responsável por uma única funcionalidade ou conjunto de funcionalidades relacionadas (microserviço?)
- Evitar dependências circulares
    - Quando um módulo depender de outros, evite que o primeiro também dependa do segundo (evitar loops de importação)
- Nomes claros e descritivos
    - Dê nomes intuitivos para os módulos e pacotes que descrevam sua função no sistema

### Projetos de maior escala
    - Estrutura hierárquica mais complexa, onde subpacotes são usados para organizar melhor o código
    

# Exemplo prático de modularização

## Ferramentas que ajudam na modularização

- requirements.txt
    - Usados para definir dependências do projeto e facilitar a distribuição
- virtualenv
    - Ferramentas para gerenciamento de ambientes virtuais e dependências, garantindo que as bibliotecas de terceiros estejam isoladas e versionadas corretamente
