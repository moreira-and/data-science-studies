PA=PB+PC-PAuPB > Probabilidade de B + Probabilidade de C - Probabilidade de B e C

Regras multiplicativas, é a multiplicação das probabilidades (e) PM = PB*PC
Probabilidade de um evento ocorrer após outro evento ter sido 
Caso o segundo evento seja afetado pelo primeiro.
PA = PB*P(C|B) > Probabilidade de B * Probabilidade de C, dado sucesso em A

Teste F para análise de variância (ANOVA)
Há situações em que podemos desejar testar a diferença entre  três  ou  mais  médias  e  já  vimos  anteriormente  que o teste t nos permite comparar apenas pares de  médias. Procedimento padrão: 
- Utilizar qq plot (quantil-quantil) do conjunto amostral em relação a distribuição normal.
- Shapiro wilk para determinar se a distribuição é normal. (H0 = Distribuição normal) (p valor>alfa, significancia com 95% de confiança)
 - Escolher testes de inferência com base na distribuição.
    - Normal: Teste de hipótese da amostra em relação a população
    - T: Teste de hipótese entre duas amostras
    - Qui-Quadrado: Teste de hipótese entre dados categóricos
