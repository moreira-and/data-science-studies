```python

# Plot da tabela de conting�ncia
sns.heatmap(tabela_contingencia, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Tabela de Conting�ncia entre Estado Civil e Possui Im�vel Pr�prio')
plt.xlabel('Possui Im�vel Pr�prio')
plt.ylabel('Estado Civil')
plt.show()
```

Hipótese Nula (H0): As duas variáveis são independentes.

Hipótese Alternativa (H1): As duas variáveis não são independentes (ou seja, estão associadas de alguma forma).

```python
# Realiza teste qui-quadrado
chi2, p, dof, expected = chi2_contingency(tabela_contingencia)
print(f"Chi2: {chi2}, p-valor: {p}, Graus de Liberdade: {dof}")
```

R: p>(1-alfa), H0 não é rejeitada. Ao p valor de 72,87%, não há evidências para rejeitar a hipótese nula de independência entre as variáveis, ou seja, o Estado Civil não tem associação significativa com o fato de o cliente ter ou não ter Imóvel Próprio.


- P-Valor Baixo:
    - Um p-valor baixo (geralmente menor que um nível de significância, como 0.05 ou 0.10) indica que a evidência contra H0 é forte. Portanto, você rejeita a hipótese nula.
    - Significado: Isso sugere que existe uma diferença significativa entre as amostras ou um efeito presente. Por exemplo, se o p-valor for 0.03, isso significa que há apenas 3% de chance de observar a estatística do teste se H0 fosse verdadeira, o que é considerado uma evidência forte contra H0.
- P-Valor Alto:
    - Um p-valor alto (geralmente maior que o nível de significância) indica que a evidência contra H0 é fraca. Nesse caso, você não rejeita a hipótese nula.
    - Significado: Isso sugere que não há diferença estatisticamente significativa entre as amostras ou que o efeito não é relevante. Por exemplo, um p-valor de 0.73, como o que você obteve, indica que não há evidências suficientes para concluir que as amostras são diferentes.
